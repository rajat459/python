```python
#$ encoding ---> transforn categorial data into nummeric value
# 1.labelencoding---> it will convert categorial data into numeric value one by one
# 2. we cannot change sub--categories encoded values
```

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/covid_toy.csv')
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
si=SimpleImputer()
```

```python
df['fever']=si.fit_transform(df[['fever']])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
df['gender']=lb.fit_transform(df['gender'])
df['cough']=lb.fit_transform(df['cough'])
df['city']=lb.fit_transform(df['city'])
df['gender']=lb.fit_transform(df['has_covid'])
```

```python
df.head(5)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(columns=['has_covid'])
y=df['has_covid']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
np.round(x_train.describe(),2)
```

```python
from sklearn.preprocessing import MinMaxScaler , StandardScaler
```

```python
sd=StandardScaler()
mn=MinMaxScaler()
```

```python
x_train_sc=sd.fit_transform(x_train)
```

```python
x_train_new=pd.DataFrame(x_train_sc,columns=x_train.columns)
```

```python
np.round(x_train_new.describe(),2)
```

```python
np.round(x_train.describe(),2)
```

```python
x_train_mn=mn.fit_transform(x_train)
```

```python
x_train_new=pd.DataFrame(x_train_mn,columns=x_train.columns)
```

```python
np.round(x_train_new.describe(),2)
```
