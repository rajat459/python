```python
import pandas as pd
# DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Edward'],
    'Age': [29, 35, 40, 30, 45],
    'Department': ['HR', 'Engineering', 'HR', 'Engineering', 'Marketing'],
    'Salary': [50000, 70000, 55000, 75000, 60000]
}
df = pd.DataFrame(data)

# 1. Calculate average salary by department
avg_salary_by_dept = df.groupby('Department')['Salary'].mean()
print("Average Salary by Department:")
print(avg_salary_by_dept)

# 2. Highest salary by department
highest_salary_by_dept = df.groupby('Department')['Salary'].max()
print("\nHighest Salary by Department:")
print(highest_salary_by_dept)

# 3. Add 'years_in_company' column
df['years_in_company'] = 2024 - 2020
print("\nData with 'years_in_company' added:")
print(df)
```

```python
data = {
    'OrderID': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    'CustomerID': ['C001', 'C002', 'C003', 'C001', 'C004', 'C005', 'C002'],
    'Product': ['Laptop', 'Phone', 'Tablet', 'Laptop', 'Phone', 'Laptop', 'Tablet'],
    'Quantity': [1, 2, 1, 1, 3, 1, 2],
    'Price': [800, 500, 300, 800, 500, 800, 300],
    'Date': ['2024-01-01', '2024-01-02', '2024-01-05', '2024-01-06', '2024-01-08', '2024-01-10', '2024-01-12'],
    'Region': ['North', 'South', 'East', 'North', 'West', 'South', 'East']
}

cf=pd.DataFrame(data)
print(cf)
```

```python
cf.isnull().sum()
```

```python
cf['total_sales']=cf['Quantity']*cf['Price']
cf
```

```python
sales_by_region = cf.groupby('Region')['total_sales'].sum()
sales_by_region
```
