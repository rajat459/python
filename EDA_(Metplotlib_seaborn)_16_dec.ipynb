```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
# Bivariate Analysis
# 1> Scatterplot(numerical column - numerical column)
```

```python
sns.scatterplot(x=tips['total_bill'],
                y=tips['tip'])
```

```python
sns.scatterplot(x='total_bill', y = 'tip',data=tips, hue = tips['sex'])
```
